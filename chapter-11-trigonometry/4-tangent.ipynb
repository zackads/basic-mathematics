## 2

Discuss how the tangent is increasing or decreasing for $-\frac{\pi}{2} \lt x \leq 0$.  Also for $\frac{\pi}{2} \lt x \lt \frac{3 \pi}{2}$ and $-\frac{3 \pi}{2} \lt x \lt - \frac{\pi}{2}$.

---

Lets define the measure of the angle $A = x$ to avoid confusion for a coordinate.

$-\frac{\pi}{2} \lt A \leq 0$ is the *fourth quadrant* of the unit circle.

Let's imagine a ray from $O$ to a point $P = (0, -1)$ on the unit circle, which creates an angle of size $A = - \frac{\pi}{2}$.

```python
circle = plt.Circle((0, 0), 1, color='black', fill=False)
fourth_quadrant = Wedge((0, 0), 1, -90, 0, hatch='/', fill=False)
P = (0, -1)

plt.gca().add_patch(circle)
plt.gca().add_patch(fourth_quadrant)
plt.gca().set_aspect('equal')
plt.xlim([-1.5, 1.5])
plt.ylim([-1.5, 1.5])
plt.plot(*P, 'o', color='black')
plt.annotate('P = (0, -1)', (P[0], P[1] - 0.3))
```

At the start of the range, $A = -\frac{\pi}{2}$, so

$$
\tan(A) = \frac{\sin(-\frac{\pi}{2})}{\cos(-\frac{\pi}{2})} = \frac{-1}{0} = \text{undefined}
$$

But if we move $P$ anticlockwise by a tiny amount by adding a _really small value_, $k$, to $A$, we get the first value of $\tan$ in the interval:

$$
\tan(-\frac{\pi}{2} + k) = \frac{\sin(-\frac{\pi}{2} + k)}{\cos(-\frac{\pi}{2} + k)} \approx \frac{-1}{k}
$$

We know from intuition that any number divided by a really small number is itself a really big number.

So can intuit that $-\frac{1}{k}$ is a really big negative number, and that's where the value of $\tan$ in the interval $-\frac{\pi}{2} \lt x \leq 0$ starts.

What happens when we keep increasing $k$?

As $k$ tends towards $\frac{\pi}{2}$, $A$ tends towards $0$ (upper limit of the interval) and

$$
\lim_{k \to  \frac{\pi}{2}} = \begin{cases}
\sin(-\frac{\pi}{2} + k) = 0 \\
\cos(-\frac{\pi}{2} + k) = 1 \\
\tan(-\frac{\pi}{2} + k) = 0
\end{cases}
$$

We can draw show this in a table

\begin{array} {|r|r|}
\hline
a & \sin a & \cos a & \tan a \\ \hline
0 & 0 & 1 & 0 \\ \hline
-\frac{\pi}{6} & -\frac{1}{2} & \frac{\sqrt{3}}{2} & -\frac{\frac{1}{2}}{\frac{\sqrt{3}}{2}} = -\frac{1}{\sqrt{3}} \\ \hline
-\frac{\pi}{4} & -\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} & -1 \\ \hline
-\frac{2 \pi}{6} = -\frac{\pi}{3} & -\frac{\sqrt{3}}{2} & \frac{1}{2} & -\sqrt{3} \\ \hline
-\frac{\pi}{2} & -1 & 0 & \frac{-1}{0} = \text{undefined} \\ \hline
\end{array}

And plot the values of the $\tan$ function next to the relevant points on the unit circle:

```python
fig, (circle_ax, tan_ax) = plt.subplots(1, 2)
tan_ax.set_ylabel('tan a')
tan_ax.set_xlabel('a')
fig.subplots_adjust(wspace=0.5)

fig.suptitle('-pi/2 < x <= 0')

circle = plt.Circle((0, 0), 1, color='black', fill=False)
A = [
    - np.pi / 2,
    - np.pi / 3,
    - np.pi / 4,
    - np.pi / 6,
    0,
]
points = [(np.cos(a), np.sin(a)) for a in A]

circle_ax.add_patch(circle)
circle_ax.set_aspect('equal')
plt.xlim([- 4 * np.pi / 2, 4 * np.pi / 2])
plt.ylim([-2, 2])
for P in points:
    circle_ax.plot(*P, 'o', color='black')
    circle_ax.plot(*zip((0, 0), P), color='black', linestyle='dashed')

for a in A:
    tan_ax.plot(*(a, np.tan(a)), 'o', color='black')

x = np.linspace(*plt.xlim())
y = np.tan(x)
tan_ax.plot(x, y, color='black', linestyle='dashed')
```

Notice how in the fourth quadrant $\tan$ is negative.

Also notice that $\tan a$ is $-1$ for $a = -\frac{\pi}{4}$, because $\sin(- \frac{\pi}{4})$ is the negation of $\cos(- \frac{\pi}{4})$.

```python
fig, (circle_ax, tan_ax) = plt.subplots(1, 2)
tan_ax.set_ylabel('tan x')
tan_ax.set_xlabel('x')
fig.subplots_adjust(wspace=0.5)

fig.suptitle('pi/2 < x < 3pi/2')

circle = plt.Circle((0, 0), 1, color='black', fill=False)
A = [
    np.pi / 2,
    2 * np.pi / 3,
    3 * np.pi / 4,
    5 * np.pi / 6,
    np.pi,
    7 * np.pi / 6,
    5 * np.pi / 4,
    4 * np.pi / 3,
    3 * np.pi / 2
]
points = [(np.cos(a), np.sin(a)) for a in A]

circle_ax.add_patch(circle)
circle_ax.set_aspect('equal')
plt.xlim([- 4 * np.pi / 2, 4 * np.pi / 2])
plt.ylim([-2, 2])
for P in points:
    circle_ax.plot(*P, 'o', color='black')
    circle_ax.plot(*zip((0, 0), P), color='black', linestyle='dashed')

for a in A:
    tan_ax.plot(*(a, np.tan(a)), 'o', color='black')

x = np.linspace(*plt.xlim())
y = np.tan(x)
tan_ax.plot(x, y, color='black', linestyle='dashed')
```

### 5

Prove that $1 + \tan^2 x = \sec^2 x$.

---

Using the Pythagorean identity $\sin^2 x + \cos^2 x \equiv 1$.

$$
\begin{align}
&1 + \tan^2 x \\
&= 1 + \frac{\sin^2 x}{\cos^2 x} \\
&= \frac{\cos^2 x}{\cos^2 x} + \frac{\sin^2 x}{\cos^2 x} \\
&= \frac{\sin^2 x + \cos^2 x}{\cos^2 x} \\
&= \frac{1}{\cos^2 x} \\
&= \sec^2 x \\
\end{align}
$$

### 6

First, a Pythagorean identity for $\operatorname{cosec}$.

$$
\operatorname{cosec} x \equiv \frac{1}{\sin}
$$

So:

$$
\begin{align}
\sin^2 x + \cos^2 x &= 1 \\
\frac{\sin^2 x}{\sin^2 x} + \frac{\cos^2 x}{\sin^2 x} &= \frac{1}{\sin^2 x} \\
1 + \cot^2 x &= \frac{1}{\sin^2 x} \\
\therefore \operatorname{cosec}^2 x &= 1 + \cot^2 x
\end{align}
$$

Then for $\cot$.

$$
\operatorname{cot} x \equiv \frac{1}{\tan} \equiv \frac{\cos}{\sin}
$$

We can re-arrange our Pythagorean identity for $\operatorname{cosec}$ in terms of $\cot$:

$$
\begin{align}
\operatorname{cosec}^2 x &\equiv 1 + \cot^2 x \\
\therefore \cot^2 x &\equiv \operatorname{cosec}^2 x - 1
\end{align}
$$
