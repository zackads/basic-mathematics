## The big ideas

### Addition

We can sum two points in a plane.

Given $A = (a_1, a_2)$ and $B = (b_1, b_2)$, we define A + B as

$$ A + B = (a_1 + b_1, a_2 + b_2) $$

The sum of two or more ordered pairs is the sum of their corresponding real numbers.

The analytic properties of adding numbers also apply to adding points: commutativity, associativity, the zero element and additive inverse are all exactly the same.

We can map the analytic addition of two ordered pairs to the geometric equivalent by plotting the operation on a plane:

```python
def add(point1, point2):
    return (point1[0] + point2[0], point1[1] + point2[1])

A = (1, 2)
B = (3, 1)
AplusB = add(A, B)
O = (0, 0)

plt.plot(*O, 'x')
plt.plot(*A, 'o')

plt.annotate('A', A)
plt.annotate('O', O)

plt.plot(*B, 'o')
plt.plot(*AplusB, 'o')

plt.annotate('B', B)
plt.annotate('A + B', AplusB)

plt.plot([A[0], AplusB[0]], [A[1], AplusB[1]])
plt.plot([O[0], B[0]], [O[1], B[1]])
plt.plot([O[0], A[0]], [O[1], A[1]])
plt.plot([B[0], AplusB[0]], [B[1], AplusB[1]])
```

### Subtraction

Subtraction is similar: (Remember, subtraction is the addition of the additive inverse: $A - B = A + (-B)$.  So in the Python below, I've defined subtraction of two points in terms of addition.)

```python
def scale(point, factor):
    return tuple(map(lambda n : n * factor, point))

def subtract(minuend, subtrahend):
    """
    Subtraction is the addition of the additive inverse,
    and the additive inverse is the scaling by -1
    """
    return add(minuend, scale(subtrahend, -1))

O = (0, 0)
A = (1, 2)
B = (3, 1)
OminusB = subtract(O, B)
AminusB = subtract(A, B)

plt.plot(*O, 'x')
plt.plot(*A, 'o')
plt.plot([O[0], B[0]], [O[1], B[1]])
plt.annotate('A', A)
plt.annotate('O', O)

plt.plot(*B, 'o')

plt.plot([A[0], AminusB[0]], [A[1], AminusB[1]])
plt.annotate('B', B)
plt.plot(*OminusB, 'o')
plt.annotate('-B', OminusB)
plt.annotate('A - B', AminusB)

plt.plot(*AminusB, 'o')
plt.plot([OminusB[0], O[0]], [OminusB[1], O[1]], '--')
plt.plot([O[0], A[0]], [O[1], A[1]], '--')
plt.plot([OminusB[0], AminusB[0]], [OminusB[1], AminusB[1]], '--')
plt.plot([A[0], B[0]], [A[1], B[1]])
plt.plot([O[0], AminusB[0]], [O[1], AminusB[1]])
```

### Translation

Addition or subtraction of one point to/from another, such that the result is a new point in the same plane, is called **translation**.  Translation can be represented symbolically with the association $P \to P + A$, abbreviated to $T_A$.

Translation is the _geometric_ operation on a _point_, whereas _addition_ is the _analytic_ operation on an _ordered pair_.  Analytic geometry shows us that they are the same thing.  Lang states this:

> We see that we have been able to define one more of the intuitive geometric notions within our system of coordinates, based only on properties of numbers.

### Distance

We can describe distance as a difference between two points, i.e. a subtraction.

By convention, we choose to use the absolute value or magnitude sign as shorthand for the distance of a point from the origin:

$$d(A, O) = |A - O| = |A|$$

(The origin is the zero point, so subtracting it has no effect.)

Look back at the previous plot of $A - B$.  See that:

$$d(A, B) = d(A - B, O) = |(A - B) - O| = |A - B|$$

Hence, the distance between $A$ and $B$, $d(A, B)$, is $|A - B|$.

$$d(P, Q) = d(T_A(P), T_A(Q))$$

So $d(T_A(P), T_A(Q)) = d(P + A, Q + A)$

Using the expression of distance in modulus notation at the bottom of p222, we can express this as

Simplifying:

$|(P + A) - (Q + A)| = |P + A - Q - A| = |P - Q|$

Which is the same as $d(P, Q)$, again by the same modulus notation on p222.

Therefore, $d(P, Q) = d(T_A(P), T_A(Q))$ and the theorem is proved.

The reflection of the centre of the circle, $P$, is the scaling of $P$ by $-1$, i.e. $R(P) = -P$.

To prove that the reflection of $S(r, P)$ is a circle, it suffices to show that the distance between the centre, $P$,  and any point on the circle, $M$, is the same in both the original and the reflection.  I.e. that:

$$d(M, P) = |M - P|$$

$$d(-M, -P) = |-M - (-P)|$$

Which we can apply the distributive property of multiplication over subtraction to show that

$$= |-M - (-P)| = |-(M - P)|$$

Which, because the modulus operator disregards that sign, we can write as follows:

$$|-(M-P)| = |M-P|$$

Which is the same as for $d(M, P)$.

We know that $d(M, P) = d(R(M), R(P))$ because $R$ is an isometry, meaning it preserves distances.

## 23

Prove that reflection through the $x$-axis is an isometry.

---

Let $P$, $Q$ be points in the plane with coordinates $(x_p, y_p)$ and $(x_q, y_q)$ respectively.

Let $R_x$ be reflection in the $x$-axis such that

$$
R_x(P) = R_x((x_p, y_p)) = (x_p, -y_p) \\
R_x(Q) = R_x((x_q, y_q)) = (xp, -y_q)
$$

$R_x$ is an isometry if $d(P, Q) = d(R_x(P), R_x(Q))$ (by the definition of an isometry as distance-preserving).

$$
d(P, Q) = d((x_p, y_p), (x_q, y_q)) \\
= \sqrt{(x_q - x_p)^2 + (y_q - y_p)^2}
$$

$$
d(R_x(P), R_x(Q) = d((x_p, -y_p), (x_q, -y_p) \\
= \sqrt{(x_q - x_p)^2 + (-y_q - (-y_p))^2} \\
= \sqrt{(x_q - x_p)^2 + ((-1)(y_q - y_p))^2} \\
= \sqrt{(x_q - x_p)^2 + (-1)^2(y_q - y_p)^2} \\
= \sqrt{(x_q - x_p)^2 + (y_q - y_p)^2}
$$

$d(P, Q) = d(R_x(P), R_x(Q))$ and so $R_x$ is an isometry.

We can express $A$ as two perpendicular transformations from the origin:

$$A = (O + (a_1, 0))) + (O + (a_2, 0)))$$

Using the definition of multiplying a point by a scalar, we can factor out $a_1$ and $a_2$ from the points $(a_1, 0)$ and $(a_2, 0)$:

$$A = (O + a_1(1, 0)) + (O + a_2(0, 1))$$

We then notice that $E_1 = (1, 0)$ and $E_2 = (0, 1)$, which we substitute into the equation:

$$A = (O + a_1E_1) + (O + a_2E_2)$$

And then we cancel the $O$s because they represent the zero point:

$$A = a_1E_1 + a_2E_2$$

which was to be proved.

b) $cE_1 = c(1, 0) = (c, 0)$. $cE_2 = c(0, 1) = (0, c)$.
