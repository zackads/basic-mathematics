### 1

Write the coordinates for $cA$ for the following values of $c$ and $A$.  In each case, draw $A$ and $cA$.

<ol type="a">
  <li>$A = (-3,5)$ and $c = 4$</li>
  <li>$A = (4,-2)$ and $c = 3$</li>
  <li>$A = (-4,-5)$ and $c = 2$</li>
  <li>$A = (2,-3)$ and $c = -2$</li>
  <li>$A = (-4,-5)$ and $c = -1$</li>
  <li>$A = (2,-3)$ and $c = 2$</li>
</ol>

```python
import matplotlib.pyplot as plt

def dilate(point, factor):
    return tuple(map(lambda n : n * factor, point))

data = [
    ("(a)", (-3, 5), 4),
    ("(b)", (4, -2), 3),
    ("(c)", (-4, -5), 2),
    ("(d)", (2, -3), -2),
    ("(e)", (-4, -5), -1),
    ("(f)", (2, -3), 2),
]

for question, A, c in data:
    plt.title(f"{question}")
    dilation = dilate(A, c)
    plt.plot(*A, 'o', color="black")
    plt.annotate(f"A {A}", A)
    plt.plot(*dilation, 'o', color="blue")
    plt.annotate(f"{c}A {dilation}", dilation)
    plt.show()
```
