## The big ideas

Analytic geometry allows us to represent points on lines and in space.  We use $\bold{R}$ to denote the set of real numbers.  If we think about this a little bit, $\bold{R}$ is just the one-dimensional real number line we learned in primary school.

If $\bold{R}$ is the one-dimensional real number **line**, then $\bold{R}^2$ is the two-dimensional **plane** of real numbers and $\bold{R}^3$ is the three-dimensional **space** of real numbers.

This is a powerful way to think about the relationship between numbers and real-world phenomena.  **Points** in the two-dimensional plane can now be represented as **ordered pairs** of one-dimensional real numbers such as $(2, -3)$.  And because they're just numbers, we can apply analytical methods to them to do powerful things.

We can multiply points by **scalar** quantities:

$$A = (a_1, a_2) \implies cA = (ca_1, ca_2)$$

For positive values of $c$, the relation $A \to cA$ has the geometric effect of stretching or **dilating** A with respect to the origin, $O$, which is left fixed:

```python
plt.xlim(-1, 6)
plt.ylim(-1, 6)

A = (1, 2)
O = (0, 0)

plt.plot(*O, 'x', color="black")
plt.annotate("O", O)
plt.plot(*A, 'o', color="black")
plt.annotate("A", A)

plt.plot(*dilate(A, 3), 'o', color="blue")
plt.annotate("3A", dilate(A, 3))
```

For values of $c$ between $0$ and $1$, $A$ is shrunk with respect to the origin:

```python
plt.plot(*O, 'x', color="black")
plt.annotate("O", O)
plt.plot(*A, 'o', color="black")
plt.annotate("A", A)

plt.plot(*dilate(A, 0.5), 'o', color="blue")
plt.annotate("0.5A", dilate(A, 0.5))
```

For values of $c < 0$, $A$ is **reflected through the origin**:

```python
plt.plot(*O, 'x', color="black")
plt.annotate("O", O)
plt.plot(*A, 'o', color="black")
plt.annotate("A", A)

plt.plot(*dilate(A, -1), 'o', color="blue")
plt.annotate("-A", dilate(A, -1))
```

### 2

Let $A$ be a point, $A \neq O$,  If $b$, $c$ are numbers such that $bA = cA$, prove that $b = c$.

---

The point $A$ is just an ordered pair of two reals $(x, y)$.

We've seen from this chapter that multiplying an ordered pair of reals by a scalar gives $c(x, y) = (cx, cy)$.

Therefore, $bA = (bx, by)$ and $cA = (cx, cy)$

$bA = cA$, so $(bx, by) = (cx, cy)$

$bA$ and $cA$ are the same points, so their $x$ and $y$ coordinates are also the same.

$bx = cx$ and $by = cy$

Cancelling the unknowns gives $b = c$ and $b = c$

Hence, $b = c$, which was to be proved.

### 3

Prove that reflection through $O$ preserves distances.  In other words, prove that $d(A, B) = d(-A, -B)$.

---

The definition of the distance between $A$ and $B$ is $d(A, B) = \sqrt{(B_x - A_x)^2 + (B_y - A_y)^2)}$.

So $d(-A, -B) = \sqrt{(-B_x - (-A)_x)^2 + (-B_y - (-A)_y)^2)}$

We can apply the distributive law for multiplication over subtraction to the two terms inside the square root:

$d(-A, -B) = \sqrt{(-(B_x - A_x))^2 + (-(B_y - A_y))^2)}$

And we know that the square of a negative number is identical to the square of the positive equivalent, so we can cancel the negation:

$d(-A, -B) = \sqrt{(B_x - A_x)^2 + (B_y - A_y)&2}$

Which leaves us with the definition of $d(A, B)$ as stated at the start.

Hence, $d(A, B) = d(-A, -B)$, which was to be proved.

c) **Theorem 1**.  Let $r$ be a positive number.  If $A$, $B$ are points, then $d(rA, rB) = r \cdot d(A, B)$.

_Proof_.  Let $A = (a_1, a_2, a_3)$ and $B = (b_1, b_2, b_3)$.  Then $rA = (ra_1, ra_2, ra_3)$ and $rB = (rb_1, rb_2, rb_3)$.  Hence

$$d(rA, rB)^2 = (rb_1 - ra_1)^2 + (rb_2 - ra_2)^2 + (rb_3 - ra_3)^2$$
$$= (r(b_1-a_1))^2 + (r(b_2-a_2))^2 + (r(b_3-a_3))^2$$
$$= r^2(b_1-a_1)^2 + r^2(b_2 - a_2)^2 + r^2(b_3-a_3)^2$$
$$= r^2 \cdot d(A, B)^2$$

Taking the square root proves our theorem.

**Theorem 2**.  Let $c$ be a number.  Then $d(cA, cB) = |c| \cdot d(A, B)$.

_Proof_.  Let $A = (a_1, a_2, a_3)$ and $B = (b_1, b_2, b_3)$.  Then $cA = (ca_1, ca_2, ca_3)$ and $rB = (cb_1, cb_2, cb_3)$.  Hence

$$d(cA, cB)^2 = (cb_1 - ca_1)^2 + (cb_2 - ca_2)^2 + (cb_3 - ca_3)^2$$
$$= (c(b_1-a_1))^2 + (c(b_2-a_2))^2 + (c(b_3-a_3))^2$$
$$= c^2(b_1-a_1)^2 + c^2(b_2 - a_2)^2 + c^2(b_3-a_3)^2$$
$$= c^2 \cdot d(A, B)^2$$

Taking the square root gives us $\sqrt{c^2}$ which, because $c$ can be any number, negative or positive, gives us the absolute value of $c$, $|c|$.
